```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load csv
nyse= pd.read_csv('NYSE.csv')
nyse
```

```python
nysedata= pd.DataFrame(nyse)
nysedata
```

```python
date= nysedata["index"]
date
```

```python
vol= nysedata["volume"]
vol
```

```python
year2019= list(date.loc[0:38807])
year2019
```

```python
vol2019= list(vol.loc[0:38807])
vol2019
```

```python
#LineGraph
plt.title('COVID-19 IMPACT ON STOCKS 2019')
plt.plot(year2019,vol2019)
plt.xlabel("2019")
plt.ylabel("Volume 2019")
```
